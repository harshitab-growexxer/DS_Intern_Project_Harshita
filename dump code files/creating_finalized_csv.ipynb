```python
import pandas as pd

# Column headers
column_headers = [
    'engine', 'cycles', 'alt', 'mach', 'TRA', 'T2', 'T24', 'T30', 'T50', 'P2', 'P15', 'P30',
    'Nf', 'Nc', 'epr', 'Ps30', 'phi', 'NRf', 'NRc', 'BPR', 'farB', 'htBleed', 'Nf_dmd', 'PCNfR_dmd',
    'W31', 'W32', 'source'
]

# Read the CSVs
train_df = pd.read_csv('/content/drive/MyDrive/turbofan dataset/train_FD002.txt', sep=' ', header=None)
test_df = pd.read_csv('/content/drive/MyDrive/turbofan dataset/test_FD002.txt', sep=' ', header=None)

train_df.drop(columns=[26, 27], inplace=True)
test_df.drop(columns=[26, 27], inplace=True)

# Add column headers
train_df.columns = column_headers[:-1]
test_df.columns = column_headers[:-1]

# Add the source column
train_df['source'] = 0
test_df['source'] = 1

# Concatenate the DataFrames
combined_df = pd.concat([train_df, test_df], ignore_index=True)

# Save the combined DataFrame to a new CSV
combined_df.to_csv('combined.csv', index=False)

print("Combined CSV created successfully.")
```
